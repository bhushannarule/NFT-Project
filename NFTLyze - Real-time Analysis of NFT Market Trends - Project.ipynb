```python
# Convert date to datetime format
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Set the date column as the Index
df.set_index('Date', inplace=True)
```

```python
plt.figure(figsize=(14,7)) 
plt.plot(df.index, df['Sales_USD_cumsum'], color='g')
plt.xlabel('Date')
plt.ylabel('Cumulative Sales (USD)')
plt.title('Cumulative NFT Sales Over Time')
plt.grid(True)
plt.show() # Plot cumulative sales over time
```

```python
plt.figure(figsize=(14, 7))
plt.hist(df['Sales_USD'].dropna(), bins=50, color='g')
plt.xlabel('Sales Price (USD)')
plt.ylabel('Frequency')
plt.title('Distribution of NFT Sales Prices')
plt.grid(True)
plt.show()  # Plot the distribution of sales prices
```

```python
import dash
from dash import dcc, html  # Updated import for Dash 2.0+
from dash.dependencies import Input, Output
import plotly.graph_objs as go
import pandas as pd  # Import pandas

app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("NFT Market Trends Dashboard"),
    dcc.Graph(id='nft-sales'),
    dcc.Interval(
        id='interval-component',
        interval=60*1000,  # Interval in milliseconds (e.g., every minute)
        n_intervals=0  # Start with 0 intervals
    )
])

@app.callback(
    Output('nft-sales', 'figure'),
    [Input('interval-component', 'n_intervals')]
)
def update_graph(n):
    df = pd.read_csv('NFTlyze Dataset.csv')  # Replace with real-time data fetching
    df['Date'] = pd.to_datetime(df['Date'])
    df.set_index('Date', inplace=True)
    df_sales = df['Sales_USD'].dropna()

    trace = go.Scatter(x=df_sales.index, y=df_sales, mode='lines', name='NFT Sales')
    return {'data': [trace], 'layout': go.Layout(title='NFT Sales Over Time')}

if __name__ == '__main__':
    app.run_server(debug=True)
```
